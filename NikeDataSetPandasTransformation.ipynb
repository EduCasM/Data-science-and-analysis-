```python
import numpy as np
import pandas as pd
```

```python
#import Data desde el csv
file= 'DataSetNike.csv'
dataset=pd.read_csv(file,
                       header=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1',
                   low_memory=False,)
dataset
```

```python
#we take only 3 variables out of 20
dataset_2= dataset.loc[:,("Material","Date","Units")]
dataset_2.head(10)
```

```python
dataset_2.shape
```

```python
#data report
def data_report(data):
    colum_names= data.columns.values
    data_types= pd.DataFrame(data.dtypes,columns=["Data_type"])
    missing_data_count= pd.DataFrame(data.isnull().sum(),columns=["missing_values"])
    unique_values_count=pd.DataFrame(columns=['unique_values'])
    
    for v in list(data.columns.values):
        unique_values_count.loc[v]=[data[v].nunique()]    
        
    data_quality_report=data_types.join(missing_data_count).join(unique_values_count).transpose()

    #quick_report= data.describe() 
    return data_quality_report
```

```python
#call data_report function
data_report(dataset_2)
```

```python
#drop rows where Material(SKU) are missing, 228 rows dropped
dataset_2=dataset_2.dropna(subset=['Material'])
```

```python
dataset_2.shape
```

```python
#drop rows where the value of units is equal or less than 0, 5561 rows are dropped
dataset_2=dataset_2[dataset_2.Units > 0]

#reset DataFrame index
dataset_2=dataset_2.reset_index(drop=True)
```

```python
dataset_2.shape
```

```python
dataset_2
```

```python
#Creating a new column named "Date_list" containing the "Date" column as a list, splitted by "/"

dataset_2=dataset_2.assign(Date_list=dataset_2['Date'].str.split("/"))
dataset_2.head(10)
```

```python
#We create new columns: "Month","Day","Year" based on the lists in the column "Date_list"
dataset_2[['Month','Day','Year']] = pd.DataFrame(dataset_2.Date_list.values.tolist())
dataset_2.head(10)
```

```python
#We eliminate the columns we will not longer need: "Date", "Date_list" and "Day"(since we will have a monthly aggregation),
# we also eliminate the hour in the "Year" column and keep only the first 4 characters of the Year string
dataset_2=dataset_2.drop(["Date","Date_list","Day"], axis=1)
dataset_2["Year"]=dataset_2.Year.str[:4]
dataset_2.head(10)
```

```python
#for convinience in the data transformation process we will separate the dataset_2 in two data sets, one for each year
# one data set for 2017 and another for 2018
#we also eliminate the "Year" column

dataset_2017= dataset_2[dataset_2.Year == "2017"].drop("Year",axis=1)
dataset_2018= dataset_2[dataset_2.Year == "2018"].drop("Year",axis=1)
```

```python
dataset_2017.shape
```

```python
dataset_2018.shape
```

```python
#we replace Month "9" to month "09" for ordering purposes 
dataset_2017["Month"].replace("9","09",inplace=True)
```

```python
#we make a group by "Material" and "Month" adding the "Units" column and then we unstack "Month" column ...
#...in order to get the desired data format.
#The months where there was not sold units we filled them with "0".Also we reset the index.
# We Change the columns names.
dataset_2017=dataset_2017.groupby(['Material', 'Month']).sum().sum(
    level=['Material', 'Month']).unstack('Month').fillna(0).reset_index()

dataset_2017.columns = ['CÓDIGO', 'SEP.', 'OCT.','NOV.','DIC.']

dataset_2018= dataset_2018.groupby(['Material', 'Month']).sum().sum(
    level=['Material', 'Month']).unstack('Month').fillna(0).reset_index()

dataset_2018.columns = ['CÓDIGO','ENE','FEB.','MAR.','ABR.','MAY.','JUN.','JUL.','AGO.','SEP.']
```

```python
dataset_2017
```

```python
dataset_2018
```

```python
#insert "AÑO" column as Int64 data type, in the "1" position 
dataset_2017.insert(1,"AÑO",2017)
dataset_2018.insert(1,"AÑO",2018)
```

```python
#Nike data types

data_report(dataset_2017)
```

```python
#Change data types to match sku(fandali) csv file 

#from float64 to int64
dataset_2017[['SEP.', 'OCT.','NOV.','DIC.']] = dataset_2017[['SEP.', 'OCT.','NOV.','DIC.']].astype(np.int64)
#from int64 to object(string)
dataset_2017[['SEP.', 'OCT.','NOV.','DIC.']] = dataset_2017[['SEP.', 'OCT.','NOV.','DIC.']].astype(str)

#from float64 to int64
dataset_2018[['ENE','FEB.','MAR.','ABR.','MAY.','JUN.','JUL.','AGO.','SEP.']] = dataset_2018[[
    'ENE','FEB.','MAR.','ABR.','MAY.','JUN.','JUL.','AGO.','SEP.']].astype(np.int64)
#from int64 to object(string)
dataset_2018[['ENE','FEB.','MAR.','ABR.','MAY.','JUN.','JUL.','AGO.','SEP.']] = dataset_2018[[
    'ENE','FEB.','MAR.','ABR.','MAY.','JUN.','JUL.','AGO.','SEP.']].astype(str)
```

```python
#adding missing columns to match  sku(fandali) csv file 
dataset_2017.insert(2,"ENE","0")
dataset_2017.insert(3,"FEB.","0")
dataset_2017.insert(4,"MAR.","0")
dataset_2017.insert(5,"ABR.","0")
dataset_2017.insert(6,"MAY.","0")
dataset_2017.insert(7,"JUN.","0")
dataset_2017.insert(8,"JUL.","0")
dataset_2017.insert(9,"AGO.","0")


dataset_2018.insert(11,"OCT.","0")
dataset_2018.insert(12,"NOV.","0")
dataset_2018.insert(13,"DIC.","0")
```

```python
#add together 2017 and 2018 data sets to get the final dataset matching the sku(fandali) csv file 
final_dataset=dataset_2017.append(dataset_2018)
```

```python
#final_dataset data report

data_report(final_dataset)
```

```python
#convertir final_dateset a CSV
final_dataset.to_csv('DataSetNike_sku.csv',encoding='latin-1',index=False)
```
